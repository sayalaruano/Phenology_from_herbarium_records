## GBIF

```python
# Load data GBIF DCA
df = pd.read_csv("GBIF_test/50species_acceptednames_and_synonyms/occurrence.txt", sep='\t', on_bad_lines='skip')
```

```python
# Create a df of GBIF DCA columns 
list_columns = pd.DataFrame(df.columns)
```

```python
# Export df
list_columns.to_csv("GBIF_test/columns_gbif_names.csv", header=False, index=False)
```

```python
# List of important columns
columns = ["gbifID", "identifier", "license", "references", "institutionCode", "collectionCode", 
           "basisOfRecord", "dynamicProperties", "catalogNumber", "recordNumber", "recordedBy",
          "lifeStage", "reproductiveCondition", "occurrenceRemarks", "eventDate", "year", "month",
          "day", "habitat", "samplingProtocol", "fieldNotes", "eventRemarks", "higherGeography",
          "countryCode", "locality", "decimalLatitude", "decimalLongitude", "scientificName",
          "family", "genus", "species", "taxonomicStatus", "elevation"]
```

```python
# Create a dataset with selected columns
df_small = df[columns]
```

```python
# Df of only rep condition (label - y variable)
rep = df_small["reproductiveCondition"]

# Drop rows without rep cond data
rep = rep.dropna()

len(rep)
```

```python
rep.unique()
```

```python
# Dfs of field notes
fn_df = df_small["occurrenceRemarks"]
fn_df = fn_df.dropna()
fn_df = df_small.loc[fn_df.index]
len(fn_df)
```

```python
fn_df2 = df_small["fieldNotes"]
fn_df2 = fn_df2.dropna()
fn_df2 = df_small.loc[fn_df2.index]
len(fn_df2)
```

```python
fn_df3 = df_small["dynamicProperties"]
fn_df3 = fn_df3.dropna()
fn_df3 = df_small.loc[fn_df3.index]
len(fn_df3)
```

```python
# Merge datasets
final_df = pd.concat([fn_df, fn_df2])
final_df = pd.concat([final_df, fn_df3])
# Delete duplicates
final_df = final_df.drop_duplicates()
```

```python
final_df
```

```python
# Df of only rep condition (label - y variable)
rep2 = final_df["reproductiveCondition"]

# Drop rows without rep cond data
rep2 = rep2.dropna()

len(rep2)
```

```python
# Merge data of images 
# Load media df
df_media = pd.read_csv("GBIF_test/Acalypha_diversifolia_gbif_media_data.txt", sep='\t')

# Select useful columns 
df_media = df_media[["gbifID", "identifier"]]

# Detele NAs
non_na = df_media["identifier"].dropna()
df_media = df_media.loc[non_na.index]

# Delete duplicates
no_dup = df_media["gbifID"].drop_duplicates()
df_media = df_media.loc[no_dup.index]

# Rename identifier column 
df_media = df_media.rename(columns={"identifier":"image_url"})
```

```python
df_media
```

```python
# Add information of media to the field notes dataset
final_df = final_df.merge(df_media, on=['gbifID'], how="left")
```

```python
final_df
```

```python
# Export final dataset
final_df.to_csv("GBIF_test/Acalypha_diversifolia_gbif_filtered_fieldnotes.csv", index=False)
```
